# Spliting Data Set

```python
X = df.drop('Class',axis = 1)
Y = df['Class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state = 42)
```

```python
x_smote, y_smote = smote.fit_resample(X,Y)
```

# Split for Train And Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_smote,y_smote,test_size=0.20,random_state=42)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 42)
model.fit(X_train,Y_train)
```

# Generating Predictions

```python
y_pred = model.predict(X_test)
```

# Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
classification_report_l = classification_report(Y_test, y_pred)
print(classification_report_l)
```

```python
confusion_matrix_l = confusion_matrix(Y_test, y_pred)
print(confusion_matrix_l)
```

```python
confusion_matrix_df = pd.DataFrame(confusion_matrix_l,columns=["Predicted Negative(0)","Predicted Positive(1)"],
                                                      index=["Actual Negative(0)","Actual Positive(1)"])
confusion_matrix_df
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(confusion_matrix_df, annot = True, fmt = "d",cmap = "Blues")
plt.title("confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
